```python
img1 = cv2.imread("img1 (1).jfif")
img2 = cv2.imread("img2 (1).jfif")


img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Resize second image to match first
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
```

```python
def show_image(title, img):
    plt.figure(figsize=(4,4))
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
show_image("Image 1", img1)
show_image("Image 2", img2)
```

```python
add = cv2.add(img1, img2)
show_image("Addition", add)
```

```python
sub = cv2.subtract(img1, img2)
show_image("Subtraction", sub)
```

```python
mul = cv2.multiply(img1, img2)
show_image("Multiplication", mul)
```

```python
div = cv2.divide(img1, img2 + 1)
show_image("Division", div)
```

```python
#logical operations
bit_and = cv2.bitwise_and(img1, img2)
show_image("Bitwise AND", bit_and)
```

```python
bit_or = cv2.bitwise_or(img1, img2)
show_image("Bitwise OR", bit_or)
```

```python
bit_xor = cv2.bitwise_xor(img1, img2)
show_image("Bitwise XOR", bit_xor)
```

```python
blend = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)
show_image("Blended Image", blend)
```

```python
abs_diff = cv2.absdiff(img1, img2)
show_image("Absolute Difference", abs_diff)
```

```python
tx = 100   # shift right (pixels)
ty = 50    # shift down (pixels)
T = np.float32([
    [1, 0, tx],
    [0, 1, ty]
])

# Perform translation
translated = cv2.warpAffine(img1, T, (img1.shape[1], img1.shape[0]))

# Display result
show_image("Image Translation", translated)
```

```python
flip_horizontal = cv2.flip(img1, 1)   # Horizontal flip
flip_vertical = cv2.flip(img1, 0)     # Vertical flip

show_image("Horizontal Flip", flip_horizontal)
show_image("Vertical Flip", flip_vertical)
```

```python
(h, w) = img1.shape[:2]
center = (w // 2, h // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img1, rotation_matrix, (w, h))

show_image("Rotated Image (45 Degrees)", rotated)
```

```python
resized = cv2.resize(img1, None, fx=0.5, fy=0.5)
show_image("Resized Image (50%)", resized)
```

```python
import numpy as np
import matplotlib.pyplot as plt

size = 400
block = size // 2

chess_2x2 = np.zeros((size, size), dtype=np.uint8)

chess_2x2[0:block, 0:block] = 255
chess_2x2[block:size, block:size] = 255

plt.imshow(chess_2x2, cmap='gray')
plt.title("2x2 Chessboard")
plt.axis('off')
plt.show()
```

```python
size = 400
diag_img = np.zeros((size, size), dtype=np.uint8)

for i in range(size):
    for j in range(size):
        if i > j:
            diag_img[i, j] = 255

plt.imshow(diag_img, cmap='gray')
plt.title("Diagonal Separation Image")
plt.axis('off')
plt.show()

```

```python
size = 400
squares = 8
block = size // squares

full_chess = np.zeros((size, size), dtype=np.uint8)

for i in range(squares):
    for j in range(squares):
        if (i + j) % 2 == 0:
            full_chess[i*block:(i+1)*block, j*block:(j+1)*block] = 255

plt.imshow(full_chess, cmap='gray')
plt.title("Full Chessboard (8x8)")
plt.axis('off')
plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

digit1 = 7
size = 200

digit_img1 = np.zeros((size, size), dtype=np.uint8)

cv2.putText(
    digit_img1,
    str(digit1),
    (40, 150),
    cv2.FONT_HERSHEY_SIMPLEX,
    5,
    255,
    10
)

plt.imshow(digit_img1, cmap='gray')
plt.title("PRN Digit 1")
plt.axis('off')
plt.show()
```

```python
digit2 = 6
size = 200

digit_img2 = np.zeros((size, size), dtype=np.uint8)

cv2.putText(
    digit_img2,
    str(digit2),
    (40, 150),
    cv2.FONT_HERSHEY_SIMPLEX,
    5,
    255,
    10
)

plt.imshow(digit_img2, cmap='gray')
plt.title("PRN Digit 2")
plt.axis('off')
plt.show()

```

```python
concat_digits = cv2.hconcat([digit_img1, digit_img2])
concat_digits=cv2.resize(concat_digits,(200,200))
plt.imshow(concat_digits, cmap='gray')
plt.title("Concatenated PRN Digits")
plt.axis('off')
plt.show()
```
