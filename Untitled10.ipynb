```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("airline_passenger_satisfaction.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[num_cols] < (Q1 - 1.5 * IQR)) |
          (df[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
extra_df = df[['ID']].copy()
extra_df['Extra_Info'] = 'Sample'

df = pd.merge(df, extra_df, on='ID', how='left')
```

```python
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

```python
scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
numeric_df = df.select_dtypes(include=np.number)
columns_with_high_variance = numeric_df.columns[numeric_df.var() > 0.01]
df = df[columns_with_high_variance]
```

```python
pca = PCA(n_components=5)
pca_data = pca.fit_transform(df[num_cols])

pca_df = pd.DataFrame(pca_data, columns=['PC1','PC2','PC3','PC4','PC5'])
pca_df.head()
```

```python
sample_df = df.sample(frac=0.3, random_state=42)
sample_df.head()
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0, 0.33, 0.66, 1.0],
    labels=['Young', 'Adult', 'Senior']
)

df[['Age', 'Age_Group']].head()
```

```python
# Make sure Age is numeric
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Drop missing values
age_data = df['Age'].dropna()
```

```python
# Number of bins
bins = 5

# Create bins
df['Age_Bin'] = pd.cut(age_data, bins=bins)
df[['Age', 'Age_Bin']].head()
```

```python
# Mean smoothing
bin_means = df.groupby('Age_Bin')['Age'].transform('mean')
df['Age_Smoothed_Mean'] = bin_means

df[['Age', 'Age_Smoothed_Mean']].head()
```

```python
# Median smoothing
bin_medians = df.groupby('Age_Bin')['Age'].transform('median')
df['Age_Smoothed_Median'] = bin_medians

df[['Age', 'Age_Smoothed_Median']].head()
```

```python
def smooth_by_boundaries(series):
    sorted_values = series.sort_values()
    min_val = sorted_values.iloc[0]
    max_val = sorted_values.iloc[-1]

    return series.apply(
        lambda x: min_val if abs(x - min_val) < abs(x - max_val) else max_val
    )

df['Age_Smoothed_Boundary'] = (
    df.groupby('Age_Bin')['Age']
      .transform(smooth_by_boundaries)
)

df[['Age', 'Age_Smoothed_Boundary']].head()
```

```python
df.to_csv("preprocessed_airline_data.csv", index=False)
files.download("preprocessed_airline_data.csv")
```
